```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load Dataset
df = pd.read_csv("Amazon.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info
```

```python
#Data Cleaning
df.drop_duplicates(inplace = True)
df.isnull().sum()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
#Core business metrics
#Total Revenue
df["Total"].sum()
```

```python
#Average Order value
df["Total"].mean()
```

```python
#Orders Count
df["Invoice ID"].nunique()
```

```python
#Category/Product Analysis
df.groupby("Product line")["Total"].sum().sort_values(ascending = True)
```

```python
#Time base analysis
df.groupby(df["Date"].dt.month)["Product line"].sum()
```

```python
#visualization
df.groupby("Product line")["Total"].sum().plot(kind = "bar")
plt.title("Category wise Revenue")
plt.show()
```

```python
df.groupby(df["Date"].dt.month)["Total"].sum().plot()
plt.title("Monthly Sales Trend")
plt.show()
```

```python
#Save outputs
df.groupby("Product line")["Total"].sum().to_csv("Sales.ipynb")
```
