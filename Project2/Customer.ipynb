```python
import pandas as pd
```

```python
#Load data
df = pd.read_csv("CustomerData.csv")
df.head()
```

```python
df.info()
```

```python
#Duplicate remove
df.drop_duplicates(inplace = True)
```

```python
df["Number"].nunique()
```

```python
df["Number"].nunique()
```

```python
df.groupby("Number")["Last_Purchase_Amount"].sum().sort_values(ascending = False).head()
```

```python
#Total customers
df["Number"].nunique()
```

```python
# total revenue
df["Last_Purchase_Amount"].sum()
```

```python
#average spend per customer
df.groupby("Number")["Last_Purchase_Amount"].sum().mean()
```

```python
#Import Graph library
import matplotlib.pyplot as plt
```

```python
df.groupby("Number")["Last_Purchase_Amount"].sum().sort_values(ascending = False).head().plot(kind = "bar")
plt.title("Top Customers by Revenue")
plt.show()
```

```python
#Key Insights:
# 1. A small number of customers contribute a large share of revenue.
# 2. Repeat customers generate higher average revenue.
# 3. Business should focus on customer retention strategies.

Revenue concentration  
I noticed that only a handful of customers are driving most of the revenue.
It’s a classic 80/20 situation where a small group contributes disproportionately to overall sales.

Value of repeat customers  
Customers who come back more often tend to spend more on average.
Their purchase frequency directly translates into higher revenue, which makes them far more valuable than one‑time buyers.

Retention focus  
The data makes it clear that keeping customers engaged for the long term is critical.
Investing in loyalty programs, personalized offers, or membership benefits would help sustain growth and profitability.
```
